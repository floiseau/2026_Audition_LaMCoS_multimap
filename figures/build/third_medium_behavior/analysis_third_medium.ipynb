# Analysis of the behavior of the third media

```python
%display latex
from sage.misc.trace import trace

# Variables
la = var("la", latex_name=r"\lambda") # Load (extension)
mu = var("mu")
ka = var("kappa")

# Values
subs = {mu : 1, ka : 1}

# Define the gradient of the transformation
F = matrix([[la, 0, 0], [0,1,0], [0,0,1]])

# Define invariants
J = det(F)
I1 = (F.T * F).trace()

# Define the energy
W = ka/2 * ln(J)^2 + mu/2 * (J^(-2/3) * I1 - 3)

# Get the stress
Pxx = diff(W, la)
display(Pxx.subs(subs))
p1 = plot(Pxx.subs(subs), (la, 1e-12, 2.5),
          figsize=(4,2),
          ymin=-10,
          axes_labels=[r"Stretch $F_{xx}$", r"Stress $P_{xx}$"])
show(p1)
save(p1, "test.svg")
```
